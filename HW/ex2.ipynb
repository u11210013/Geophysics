```python
import pygmt
import pandas as pd

# 
data = {
    "longitude": [121.5, 122.0, 121.8],
    "latitude": [25.0, 25.5, 25.2],
    "depth": [10, 20, 30]
}
df = pd.DataFrame(data)

# 
fig = pygmt.Figure()

# 
fig.coast(
    region=[119, 123, 20, 25],  # 
    projection="M6i",           # Mercator 6 
    land="gray",                # 
    water="lightblue",          # 
    borders=[1],                # 
    frame=["a", "+t'Taiwan'"], # 
)

# 
fig.plot(
    x=df["longitude"],
    y=df["latitude"],
    style="c0.5c",       # 
    color="red",        # 
    pen="black",        # 
    label="Earthquake"
)

# 
fig.show()

# 
fig.savefig("earthquake_map.png")
```

```python
import pygmt
# Create a PyGMT figure object
fig = pygmt.Figure()

# Define the region for Taiwan (longitude and latitude bounds)
region = [119, 123, 20, 25]

# Add topography data
fig.grdcontour(
    grid="@earth_relief_03m",  # GMT's global topography data at 3 arc-minute resolution
    region=region,
    projection="M6i",  # Mercator projection, 6 inches wide
    contour="1+e/2p,black",  # Draw contour lines every 1 unit
    frame=["a", "+t'Taiwan Topography'"],  # Frame with title
)

# Add coastlines
fig.coast(
    shorelines="1/0.5p,black",  # Shorelines with 0.5-point width
    water="lightblue",  # Water color
    land="gray",  # Land color
)

# Display the map
fig.show()
```

```python
import pygmt
#  PyGMT 
fig = pygmt.Figure()

# 
region = [119, 123, 20, 26]  # 

grid = pygmt.datasets.load_earth_relief(resolution="03m",region=region)

# 
fig.grdimage(
    grid=grid,  # GMT  3 
    region=region,
    projection="M6i",          # Mercator  6 
    frame=True,  # 
    shading=True
)

# 
fig.coast(
    region=region,
    borders=[1],
    shorelines=True,  #  0.5 
    water="skyblue"          # 
)

fig.colorbar(frame='af+1"Topography (m)"')
# 
fig.show()
```

```python
import pygmt

#  PyGMT 
fig = pygmt.Figure()

# 
region = [119, 123, 20, 26]  # 

# 
pygmt.makecpt(
    cmap="orange,red",  # 
    series=[-5000, 5000],  # 
    continuous=True  # 
)

# 
fig.grdimage(
    grid="@earth_relief_03m",  # GMT  3 
    region=region,
    projection="M6i",          # Mercator  6 
    cmap=True,                 # 
    frame=["a", "+t'Taiwan Topography'"],  # 
)

# 
fig.coast(
    shorelines="1/0.5p,black",  #  0.5 
    water="skyblue",          # 
)

# 
fig.show()
```
