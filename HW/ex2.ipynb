```python
import pygmt
import pandas as pd

# 示例地震數據（經度、緯度和震中深度）
data = {
    "longitude": [121.5, 122.0, 121.8],
    "latitude": [25.0, 25.5, 25.2],
    "depth": [10, 20, 30]
}
df = pd.DataFrame(data)

# 創建一個地圖對象
fig = pygmt.Figure()

# 繪製地震事件的地理位置
fig.coast(
    region=[119, 123, 20, 25],  # 台灣及其周邊的經緯度範圍
    projection="M6i",           # Mercator 投影，6 英寸寬
    land="gray",                # 陸地設置為灰色
    water="lightblue",          # 海洋設置為淡藍色
    borders=[1],                # 繪製國界
    frame=["a", "+t'Taiwan'"], # 顯示圖框和標題
)

# 繪製地震事件位置
fig.plot(
    x=df["longitude"],
    y=df["latitude"],
    style="c0.5c",       # 圓形標記
    color="red",        # 標記顏色
    pen="black",        # 標記邊框顏色
    label="Earthquake"
)

# 顯示地圖
fig.show()

# 保存地圖到文件
fig.savefig("earthquake_map.png")
```

```python
import pygmt
# Create a PyGMT figure object
fig = pygmt.Figure()

# Define the region for Taiwan (longitude and latitude bounds)
region = [119, 123, 20, 25]

# Add topography data
fig.grdcontour(
    grid="@earth_relief_03m",  # GMT's global topography data at 3 arc-minute resolution
    region=region,
    projection="M6i",  # Mercator projection, 6 inches wide
    contour="1+e/2p,black",  # Draw contour lines every 1 unit
    frame=["a", "+t'Taiwan Topography'"],  # Frame with title
)

# Add coastlines
fig.coast(
    shorelines="1/0.5p,black",  # Shorelines with 0.5-point width
    water="lightblue",  # Water color
    land="gray",  # Land color
)

# Display the map
fig.show()
```

```python
import pygmt
# 創建 PyGMT 圖形對象
fig = pygmt.Figure()

# 定義台灣的範圍
region = [119, 123, 20, 26]  # 台灣的經緯度範圍

grid = pygmt.datasets.load_earth_relief(resolution="03m",region=region)

# 使用彩色漸層來顯示地形
fig.grdimage(
    grid=grid,  # GMT 提供的全球地形數據，分辨率為 3 分鐘
    region=region,
    projection="M6i",          # Mercator 投影，寬 6 英寸
    frame=True,  # 設置圖框和標題
    shading=True
)

# 添加海岸線（不設置陸地顏色）
fig.coast(
    region=region,
    borders=[1],
    shorelines=True,  # 海岸線，線寬 0.5 點
    water="skyblue"          # 海洋顏色
)

fig.colorbar(frame='af+1"Topography (m)"')
# 顯示地圖
fig.show()
```

```python
import pygmt

# 創建 PyGMT 圖形對象
fig = pygmt.Figure()

# 定義台灣的範圍
region = [119, 123, 20, 26]  # 台灣的經緯度範圍

# 定義橘色漸層色彩映射（從淺橘色到深橘色）
pygmt.makecpt(
    cmap="orange,red",  # 使用內建的橘色漸層
    series=[-5000, 5000],  # 高程範圍，單位為米
    continuous=True  # 使用連續漸層
)

# 使用橘色漸層來顯示地形
fig.grdimage(
    grid="@earth_relief_03m",  # GMT 提供的全球地形數據，分辨率為 3 分鐘
    region=region,
    projection="M6i",          # Mercator 投影，寬 6 英寸
    cmap=True,                 # 使用自定義的橘色漸層色彩映射
    frame=["a", "+t'Taiwan Topography'"],  # 設置圖框和標題
)

# 添加海岸線
fig.coast(
    shorelines="1/0.5p,black",  # 海岸線，線寬 0.5 點
    water="skyblue",          # 海洋顏色
)

# 顯示地圖
fig.show()
```

```python
import pygmt

# 創建 PyGMT 圖形對象
fig = pygmt.Figure()

# 定義台灣的範圍
region = [119, 123, 20, 26]  # 台灣的經緯度範圍

# 使用 plasma 顏色漸層，並擴大高度範圍來增強顏色變化

pygmt.makecpt(
    cmap="plasma",             # 使用 plasma 色彩映射
    series=[-7000, 5000],      # 擴展地形高度範圍，增加對比度
    continuous=True            # 保持顏色為連續漸層
)

# 使用自定義的顏色分層來顯示地形
fig.grdimage(
    grid="@earth_relief_01m",  # GMT 提供的全球地形數據，分辨率為 3 分鐘
    region=region,
    projection="M6i",          # Mercator 投影，寬 6 英寸
    cmap=True,                 # 使用自定義的顏色漸層
    shading=True,              # 加入陰影效果來強化地形細節
    frame=["a", "+t'Taiwan Topography'"],  # 設置圖框和標題
)

# 使用預設的海岸線樣式
fig.coast(
    shorelines="1/0.5p,black",  # 顯示黑色海岸線
    water="lightblue",          # 海洋顏色
)

# 顯示地圖
fig.show()
```
